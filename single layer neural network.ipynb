```python
import pandas as pd
from datetime import datetime
import numpy as np
from tensorflow import keras
from tensorflow.keras import layers
# Load the TensorBoard notebook extension
%load_ext tensorboard
import tensorflow as tf
import datetime
from keras.callbacks import EarlyStopping

stopper = EarlyStopping(patience = 15)

# Clear any logs from previous runs
!rm -rf ./logs/ 

from kerastuner.tuners import RandomSearch

train_df = pd.read_csv('data/features.csv')

train_df = train_df[train_df.metroscubiertos <= train_df.metrostotales]

train_x = train_df.drop(columns=['precio'])
train_y = train_df.precio

n_cols = train_x.shape[0]

train_x = train_x.to_numpy()
train_y = train_y.to_numpy()
#train_x.head()
train_x.shape
```

```python
model = keras.Sequential()
model.add(layers.Dense(units = 1000, activation = 'relu', input_shape = (45,)))
model.add(layers.Dropout(rate = 0.4))
model.add(layers.Dense(units = 1))
```

```python
model.compile(optimizer = 'adam', loss = 'mean_squared_error', metrics = ['mae'])
```

```python
log_dir="logs/fit/single_layer_nn/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir = log_dir, histogram_freq=1)
```

```python
model.fit(x = train_x, y = train_y, epochs = 300, validation_split = 0.4, callbacks = [tensorboard_callback, stopper])
```

```python
test_df = pd.read_csv('data/testModified.csv')
prediction = model.predict(test_df.drop(columns=['id']))

res = pd.DataFrame(prediction, index=test_df.id, columns=['precio'])
res = res.rename({'precio':'target'}, axis=1)
res.to_csv("predictions/single_layer_prediction.csv", header = True)
```
